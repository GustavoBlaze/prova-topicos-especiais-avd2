```python
import cv2 as cv2
from google.colab.patches import cv2_imshow
import dlib
```

```python
# Carregando detector de faces e marco faciais
face_detector = dlib.get_frontal_face_detector()
point_detector = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

# Carregando imagem
image = cv2.imread('example.jpg')

# Detectando faces
faces = face_detector(image, 1)

# Loop para iterar o array de faces
for face in faces:
  # Detectando os marcos faciais de cada face encontrada
  points = point_detector(image, face)

  # Desenhando os marcos faciais de cada face encontrada
  for point in points.parts():
      cv2.circle(
          image,
          (point.x, point.y),
          2,
          (0, 255, 0),
          1
      );

  # Desenhando um retangulo em cada rosto encontrado na imagem
  cv2.rectangle(
      image,
      (face.left(), face.top()),
      (face.right(), face.bottom()),
      (113, 89, 193),
      4
  )

# Exibindo a imagem
cv2_imshow(image)
```
